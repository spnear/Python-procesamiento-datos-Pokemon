# Graficos de comparacion

```python
#             Eje x    Eje y   
pokemon.plot("Type 1","Speed")
```

```python
#Comparar cuantitativamente dos referencias (datos)
pokemon["Total"].plot()
pokemon["HP"].plot(secondary_y=True, style="y")
```

# Scatter Plots

```python
#               Eje x  Eje y     Tipo 
pokemon.plot(x="Total",y="HP",kind="scatter")
```

```python
#Dos graficas de dispersion en una sola
ax = pokemon.plot(x="Total",y="HP",kind="scatter",color="#c219D6",label="Grupo 1")
pokemon.plot(x="Attack",y="Speed",kind="scatter",color="#13E890",label="Grupo 2",ax=ax)
```

```python
#Grafico de regresion lineal 
sns.regplot(x="Total",y="HP",data=pokemon)
```

# Kernel Density Estimation 

```python
#Area bajo la curva 
sns.kdeplot(pokemon["HP"],shade=True)
```

```python
sns.distplot(pokemon["Total"],color="red")
```

# Violin Plots

#Muy parecidos a los boxplot

```python
sns.violinplot(pokemon["Total"],color="red")
```

```python
sns.violinplot(data=pokemon.drop(["#","Name","Legendary","Generation", "Total"],axis=1).head())
#Muestra media,mediana y otras cosas...  
```

```python
sns.kdeplot(pokemon["HP"],cumulative=True)
#Coge la menor y de ah empieza a estabilizarce 
```

# Join Plots

```python
#Une varios graficos en uno 
sns.jointplot(x="Total",y="Speed",data=pokemon)
```

```python
sns.jointplot(x="Total",y="Speed",data=pokemon,kind="scatter",color="purple")
```

```python
sns.jointplot(x="Total",y="Speed",data=pokemon,kind="reg",color="green")
    #kind="reg" : Regresion
```

```python
sns.jointplot(x="Total",y="Speed",data=pokemon,kind="resid",color="orange")
#Latentes y residuales
```

```python
sns.jointplot(x="Total",y="Speed",data=pokemon,kind="kde",color="red")
#Correlacion fuerte de los datos: grado de vecindad tiende a infinito, muy juntos---Lo negro
```

```python
sns.jointplot(x="Total",y="Speed",data=pokemon,kind="hex",color="blue")
#Correlacion mediante un indice de dispersion diferente al de arriba
```
